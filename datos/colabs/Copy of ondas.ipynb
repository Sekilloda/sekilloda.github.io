```python
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

mpl.rcParams['legend.fontsize'] = 10

fig = plt.figure()
ax = fig.gca(projection='3d')
#defino el parametro

#ver que no sea mas que c
f = input('Ingrese la frecuencia, en Hertz')
w = 2*np.pi*float(f)
longitud_de_onda = input('Ingrese la longitud de onda, en metros')
k = 2*np.pi/float(longitud_de_onda)
amplitud= float(input('Ingrese la amplitud, en voltios sobre metro'))
(dir_x, dir_y, dir_z) = (input('Ingrese una dirección de propagación, componente a componente'), input(), input())
dir_x=float(dir_x)
dir_y=float(dir_y)
dir_z=float(dir_z)
#normalizamos la dirección
dir_x=dir_x/np.sqrt(dir_x**2+dir_y**2+dir_z**2)
dir_y=dir_x/np.sqrt(dir_x**2+dir_y**2+dir_z**2)
dir_z=dir_x/np.sqrt(dir_x**2+dir_y**2+dir_z**2)
#sacamos proyecciones de k en cada componente
k_x=k*dir_x
k_y=k*dir_y
k_z=k*dir_z
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
#especifico la "trayectoria" x(r,theta),y(r, theta),z o w(vu)
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 100)
z = np.linspace(0, 10, 100)
t = np.linspace(0, 10, 100)
phase_x=0
phase_y=0
phase_z=0
E1 = amplitud * np.cos(k_x*x-w*t+phase_x)
E2 = amplitud * np.cos(k_y*y-w*t+phase_y)
E3 = amplitud * np.cos(k_z*z-w*t+phase_z)
ax.plot(x, E1, label='curva 1')
#ax.plot(y, E2, label='curva 2')
ax.legend()
plt.xlabel("x")
plt.ylabel("t")
plt.show()
```
